```python
def _cosine_anneal_schedule(t, T = 50, M = 5, alpha_zero = 0.02):
        cos_inner = np.pi * (t % (T // M))  # t - 1 is used when t has 1-based indexing.
        cos_inner /= T // M
        cos_out = np.cos(cos_inner) + 1
        return float(alpha_zero / 2 * cos_out)
```

```python
lr = [_cosine_anneal_schedule(i, M = 5) for i in range(50)]
```
